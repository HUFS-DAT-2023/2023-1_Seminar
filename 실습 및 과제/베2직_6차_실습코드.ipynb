## COVID-19 백신 종류별 접종 인구

```python
days=[1,2,3] # 1일, 2일, 3일
az=[2,4,8] #단위 만명
pfizer=[5,1,3]
moderna=[1,2,5]

plt.plot(days,az)
plt.plot(days, pfizer)
plt.plot(days,moderna)
```

```python
plt.plot(days,az, label='az')
plt.plot(days, pfizer, label='pfizer', marker='o', linestyle='--')
plt.plot(days,moderna, label='moderna', marker='s', ls='-.')

plt.legend()
```

```python
plt.plot(days,az, label='az')
plt.plot(days, pfizer, label='pfizer', marker='o', linestyle='--')
plt.plot(days,moderna, label='moderna', marker='s', ls='-.')

plt.legend(ncol=3)
```

# 27. 막대 그래프 (심화) 

```python
labels=['강백호', '서태웅', '정대만']
values=[190, 187, 184]

plt.barh(labels, values)
```

```python
labels=['강백호', '서태웅', '정대만']
values=[190, 187, 184]

plt.barh(labels, values)
plt.xlim(175,195)
```

```python
bar=plt.bar(labels, values)
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

```python
bar=plt.bar(labels, values)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height(), values[idx])
```

```python
bar=plt.bar(labels, values)
plt.ylim(175,195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height()+0.5, values[idx], ha='center', color='blue')
```
