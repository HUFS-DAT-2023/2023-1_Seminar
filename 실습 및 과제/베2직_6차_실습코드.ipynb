```python
import pandas as pd
df=pd.read_excel('1818.xlsx', sheet_name=1)
df
```

```python
df=pd.read_excel('1818.xlsx', skiprows = 4, skipfooter = 2)
df
```

```python
df['시도'].replace({'부산': '경상'}, inplace=True)
df
```

```python
df['시도'].replace({'대구': '경상'}, inplace=True)
df
```

```python
df['시도'].replace({'울산': '경상'}, inplace=True)
df
```

```python
df['시도'].replace({'인천': '경기', '대전':'충청', '세종':'충청', '광주': '전라', '충북':'충청','충남':'충청', '경북':'경상', '경남':'경상', '전북':'전라', '전남':'전라'}, inplace=True)
df
```

```python
df.drop(columns=['도시'], inplace=True)
df
```

```python
df=df.astype(str).apply(lambda x: x.str.replace('*', ' ').str.strip())
df
```

```python
df1 = df.astype({'1월':'float', '2월':'float', '3월':'float', '4월':'float', '5월':'float', '6월':'float', '7월':'float', '8월':'float', '9월':'float', '10월':'float', '11월':'float', '12월':'float', '연평균':'float' })
df1
```

```python
 df2= df1.groupby('시도').mean()
df2
```

```python
df3=df2.sort_values('연평균',ascending=False)
df3
```

```python
df3.to_csv('18년 NO2 대기오염.csv', encoding='utf-8-sig')
```
