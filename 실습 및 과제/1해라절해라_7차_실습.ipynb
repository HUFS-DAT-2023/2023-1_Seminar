```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus']= False
```

```python
file_path = 'boston.csv'

# CSV 파일 불러오기
df = pd.read_csv(file_path)

df
```

```python
plt.bar(df['Player'],df['Age'],label='Age')
plt.bar(df['Player'],df['G'], bottom=df['Age'],label='G')
plt.bar(df['Player'],df['GS'], bottom=df['Age']+df['G'], label='GS')

plt.xticks(rotation=60)
plt.legend()
```

```python
index = np.arange(N)
index

plt.figure(figsize=(10,5))
plt.title('수비')

w = 0.25
plt.bar(index-w,df['TRB'], width=w,label='TRB')
plt.bar(index, df['STL'],width=w, label='STL')
plt.bar(index+w, df['BLK'],width=w, label='BLK')
plt.legend(ncol=3)
plt.xticks(index, df['Player'],rotation=60)
plt.show
```

```python
values = [30,25,20,13,10,2]
labels = ['JT', 'JB', 'MS', 'AH', 'DW', 'MB']
explode=[0.05]*6
plt.pie(values, labels=labels,explode=explode)
plt.show()
```

```python
values = [30,25,20,13,10,2]
labels = ['JT', 'JB', 'MS', 'AH', 'DW', 'MB']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']
explode = [0.05]*6
plt.pie(values, labels=labels,autopct='%.1f',startangle=90,counterclock=False)
plt.show
```

```python
wedgeprops={'width':0.6,'edgecolor':'w','linewidth':2}
def custom_autopct(pct):
    return'{:0f}%'.format(pct) if pct >= 10 else '' 

plt.pie(values, labels=labels,autopct=custom_autopct,startangle=90,counterclock=False, wedgeprops=wedgeprops, pctdistance=0.7)
plt.show()
```
