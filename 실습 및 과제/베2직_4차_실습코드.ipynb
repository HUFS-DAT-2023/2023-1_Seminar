# 11. 데이터 선택(조건)
조건에 해당하는 데이터 선택

```python
df['키'] >=185 #학생들의 키가 185이상인지 여부를 T/F
```

```python
filt=(df['키'] >= 185)
df[filt]
```

```python
df[~filt] #filter를 역으로 적용
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키']>=185, '수학']  #키가 185이상인 학생들의 수학 데이터
```

```python
df.loc[df['키']>=185, ['이름', '수학', '과학']]
```

## 다양한 조건
& 그리고

```python
df.loc[(df['키']>=185) & (df['학교']=='북산고')]
```

### | 또는

```python
df.loc[(df['키']< 170) | (df['키']>200)]
```

# 12. 데이터 선택(조건) #2
### str 함수

```python
filt=df['이름'].str.startswith('송')  #송씨 성을 가진 사람
df[filt]
```

```python
filt=df['이름'].str.contains('태')  #이름에 태가 들어가는 사람
df[filt]
```

```python
df[~filt] #이름에 태가 없는 사람
```

```python
langs=["Python", "Java"]
filt=df['SW특기'].isin(langs)
df[filt]
```

## 데이터 제외하기 dropna

```python
import pandas as pd
df=pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.dropna(inplace=True) #전체 데이터 중에서 NaN을 포함하는 데이터 삭제
```

```python
df
```

```python
import pandas as pd
df=pd.read_excel('score.xlsx', index_col='지원번호')
df
```

- axis : index or column
- how : any or all

```python
df.dropna(axis='index', how='any') #NaN가 하나라도 있는 row 삭제
```

```python
df.dropna(axis='columns') #NaN가 하나라도 있는 column삭제
```

```python
df['학교']=na.nan
df
```

```python
df.dropna(axis='columns', how='all') #데이터 전체가 NaN인 경우에만 column 삭제
```
